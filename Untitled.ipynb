```python
from sklearn.impute import SimpleImputer

new_df = df.drop(['City','Datetime'],axis=1)

imputer = SimpleImputer(strategy="median")
imputer.fit(new_df)
X = imputer.transform(new_df)
imputer.statistics_
```

```python
df_tr = pd.DataFrame(X,columns=new_df.columns)
df_tr.insert(loc=0, column='City', value=df['City'])
df_tr.insert(loc=1, column='Datetime', value=df['Datetime'])
df_tr.head()
```

```python
def get_SO2_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 380:
        return 100 + (x - 80) * 100 / 300
    elif x <= 800:
        return 200 + (x - 380) * 100 / 420
    elif x <= 1600:
        return 300 + (x - 800) * 100 / 800
    elif x > 1600:
        return 400 + (x - 1600) * 100 / 800
    else:
        return 0

df_tr["SO2i"] = df_tr["SO2"].apply(lambda x: get_SO2_subindex(x))
data = df_tr[["SO2i","SO2"]]
data.head()
```

```python
## PM2.5 Sub-Index calculation
def get_PM25_subindex(x):
    if x <= 30:
        return x * 50 / 30
    elif x <= 60:
        return 50 + (x - 30) * 50 / 30
    elif x <= 90:
        return 100 + (x - 60) * 100 / 30
    elif x <= 120:
        return 200 + (x - 90) * 100 / 30
    elif x <= 250:
        return 300 + (x - 120) * 100 / 130
    elif x > 250:
        return 400 + (x - 250) * 100 / 130
    else:
        return 0

df_tr["PM2.5i"] = df_tr["PM2.5"].apply(lambda x: get_PM25_subindex(x))
data = df_tr[["PM2.5i","PM2.5"]]
data.head()
```

```python
## PM10 Sub-Index calculation
def get_PM10_subindex(x):
    if x <= 50:
        return x
    elif x <= 100:
        return x
    elif x <= 250:
        return 100 + (x - 100) * 100 / 150
    elif x <= 350:
        return 200 + (x - 250)
    elif x <= 430:
        return 300 + (x - 350) * 100 / 80
    elif x > 430:
        return 400 + (x - 430) * 100 / 80
    else:
        return 0

df_tr["PM10i"] = df_tr["PM10"].apply(lambda x: get_PM10_subindex(x))
data = df_tr[["PM10i","PM10"]]
data.head()
```

```python
## NH3 Sub-Index calculation
def get_NH3_subindex(x):
    if x <= 200:
        return x * 50 / 200
    elif x <= 400:
        return 50 + (x - 200) * 50 / 200
    elif x <= 800:
        return 100 + (x - 400) * 100 / 400
    elif x <= 1200:
        return 200 + (x - 800) * 100 / 400
    elif x <= 1800:
        return 300 + (x - 1200) * 100 / 600
    elif x > 1800:
        return 400 + (x - 1800) * 100 / 600
    else:
        return 0

df_tr["NH3i"] = df_tr["NH3"].apply(lambda x: get_NH3_subindex(x))
data = df_tr[["NH3i","NH3"]]
data.head()
```

```python
## CO Sub-Index calculation
def get_CO_subindex(x):
    if x <= 1:
        return x * 50 / 1
    elif x <= 2:
        return 50 + (x - 1) * 50 / 1
    elif x <= 10:
        return 100 + (x - 2) * 100 / 8
    elif x <= 17:
        return 200 + (x - 10) * 100 / 7
    elif x <= 34:
        return 300 + (x - 17) * 100 / 17
    elif x > 34:
        return 400 + (x - 34) * 100 / 17
    else:
        return 0

df_tr["COi"] = df_tr["CO"].apply(lambda x: get_CO_subindex(x))
data = df_tr[["COi","CO"]]
data.head()
```

```python
## O3 Sub-Index calculation
def get_O3_subindex(x):
    if x <= 50:
        return x * 50 / 50
    elif x <= 100:
        return 50 + (x - 50) * 50 / 50
    elif x <= 168:
        return 100 + (x - 100) * 100 / 68
    elif x <= 208:
        return 200 + (x - 168) * 100 / 40
    elif x <= 748:
        return 300 + (x - 208) * 100 / 539
    elif x > 748:
        return 400 + (x - 400) * 100 / 539
    else:
        return 0

df_tr["O3i"] = df_tr["O3"].apply(lambda x: get_O3_subindex(x))
data = df_tr[["O3i","O3"]]
data.head()
```

```python
## NOx Sub-Index calculation
def get_NOx_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 180:
        return 100 + (x - 80) * 100 / 100
    elif x <= 280:
        return 200 + (x - 180) * 100 / 100
    elif x <= 400:
        return 300 + (x - 280) * 100 / 120
    elif x > 400:
        return 400 + (x - 400) * 100 / 120
    else:
        return 0

df_tr["NOxi"] = df_tr["NOx"].apply(lambda x: get_NOx_subindex(x))
data = df_tr[["NOxi","NOx"]]
data.head()
```

```python
data = df_tr[["SO2i","NOxi","COi","O3i","PM10i","PM2.5i","NH3i"]]
data.head()
```

```python
## AQI bucketing
def get_AQI_bucket(x):
    if x <= 50:
        return "Good"
    elif x <= 100:
        return "Satisfactory"
    elif x <= 200:
        return "Moderate"
    elif x <= 300:
        return "Poor"
    elif x <= 400:
        return "Very Poor"
    elif x > 400:
        return "Severe"
    else:
        return np.NaN

df_tr["Checks"] = (df_tr["PM2.5i"] > 0).astype(int) + \
                (df_tr["PM10i"] > 0).astype(int) + \
                (df_tr["SO2i"] > 0).astype(int) + \
                (df_tr["NOxi"] > 0).astype(int) + \
                (df_tr["NH3i"] > 0).astype(int) + \
                (df_tr["COi"] > 0).astype(int) + \
                (df_tr["O3i"] > 0).astype(int)

df_tr["AQI"] = round(df_tr[["PM2.5i", "PM10i", "SO2i", "NOxi",
                                 "NH3i", "COi", "O3i"]].max(axis = 1))
df_tr.loc[df_tr["PM2.5i"] + df_tr["PM10i"] <= 0, "AQI"] = np.NaN
df_tr.loc[df_tr.Checks < 3, "AQI"] = np.NaN

df_tr["AQI_Range"] = df_tr["AQI"].apply(lambda x: get_AQI_bucket(x))
# df_tr[~df_tr.AQI.isna()].head(13)
```

```python
data = df_tr[["PM2.5","PM10","SO2","NOx","NH3","CO","O3","AQI","AQI_Range"]]
data.head()
```

```python
X = df_tr[["PM2.5i","PM10i","SO2i","NOxi","NH3i","COi","O3i"]]
Y = df_tr[["AQI_Range"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
print(len(X_train),len(Y_train),len(X_test),len(Y_test))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier().fit(X_train,Y_train)
predict_Ytrain = model.predict(X_train)
predict_Ytest = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Model accuracy on training dataset : ",accuracy_score(Y_train,predict_Ytrain))
print("Model accuracy on testing dataset : ",accuracy_score(Y_test,predict_Ytest))
```
